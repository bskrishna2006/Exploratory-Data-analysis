```python
# In duration feature split the hour and the min

# Check for any unusual formats first
print("All unique Duration formats:")
unique_durations = df['Duration'].unique()
print(f"Total unique formats: {len(unique_durations)}")
for i, duration in enumerate(unique_durations[:30]):  # Show first 30
    print(f"{i+1}: '{duration}'")

# Function to extract hours
def extract_hours(duration):
    if 'h' in duration:
        return int(duration.split('h')[0])
    else:
        return 0  # If no hours, return 0

# Function to extract minutes  
def extract_minutes(duration):
    if 'm' in duration:
        if 'h' in duration:
            # Format like "2h 50m"
            parts = duration.split(' ')
            for part in parts:
                if 'm' in part:
                    return int(part.replace('m', ''))
        else:
            # Format like "5m" (minutes only)
            return int(duration.replace('m', ''))
    return 0  # If no minutes found

# Apply the functions
df['Total_Duration_Hours'] = df['Duration'].apply(extract_hours)
df['Total_Duration_Minutes'] = df['Duration'].apply(extract_minutes)

# Display the results
print("\nSample of extracted duration:")
print(df[['Duration', 'Total_Duration_Hours', 'Total_Duration_Minutes']].head(15))
```

```python
# Verify the parsing worked correctly and create total duration in minutes
print("Verification of Duration parsing:")
print("=" * 50)

# Check for any parsing errors
print(f"Original Duration values: {df['Duration'].notna().sum()}")
print(f"Hours extracted: {df['Total_Duration_Hours'].notna().sum()}")
print(f"Minutes extracted: {df['Total_Duration_Minutes'].notna().sum()}")

# Create total duration in minutes for easier analysis
df['Total_Duration_Minutes_Total'] = (df['Total_Duration_Hours'] * 60) + df['Total_Duration_Minutes']

# Show some examples
print("\nSample comparisons:")
sample_df = df[['Duration', 'Total_Duration_Hours', 'Total_Duration_Minutes', 'Total_Duration_Minutes_Total']].head(10)
print(sample_df)

# Check data types
print(f"\nData types:")
print(f"Total_Duration_Hours: {df['Total_Duration_Hours'].dtype}")
print(f"Total_Duration_Minutes: {df['Total_Duration_Minutes'].dtype}")
print(f"Total_Duration_Minutes_Total: {df['Total_Duration_Minutes_Total'].dtype}")

# Now we can drop the original Duration column
df.drop('Duration', axis=1, inplace=True)
print(f"\nDropped original Duration column. Current columns: {len(df.columns)}")
```
